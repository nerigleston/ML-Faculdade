```python
# Modelo não supervisionado de redução
# Importar as bibliotecas necessárias
import pandas as pd
from sklearn.decomposition import PCA

# Carregar os dados
df = pd.read_csv('creditcard.csv')

# Pré-processar os dados
df = df.drop(['Class'], axis=1)

# Aplicar PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(df)

# Criar DataFrame para os componentes principais
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])

print(principalDf.head())

import matplotlib.pyplot as plt

# Plotar os componentes principais em um gráfico de dispersão 2D
plt.figure(figsize=(8, 6))
plt.scatter(principalDf['principal component 1'], principalDf['principal component 2'], alpha=0.5)
plt.title('PCA Plot - Componentes Principais')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.grid(True)
plt.show()
```

A Análise de Componentes Principais (PCA) é uma técnica estatística utilizada para reduzir a dimensionalidade dos dados enquanto mantém o máximo possível de informação. Ao aplicar PCA ao conjunto de dados creditcard.csv, os valores resultantes apresentam as coordenadas das amostras em um novo espaço definido pelos componentes principais. Esses componentes principais são combinações lineares das variáveis originais, onde o primeiro componente captura a maior variabilidade dos dados e cada componente subsequente captura a variabilidade restante de forma ortogonal aos anteriores. Portanto, os valores exibidos para "principal component 1" e "principal component 2" representam as coordenadas das amostras nesse novo espaço de duas dimensões, proporcionando uma representação mais compacta e informativa dos dados originais.

A redução de dimensionalidade com PCA é importante porque simplifica a análise de dados complexos, tornando mais fácil visualizar e interpretar informações significativas. Ao reduzir os dados a poucas componentes principais, elimina-se o ruído e melhora-se o desempenho computacional, facilitando a identificação de padrões e anomalias. Além disso, ajuda a prevenir overfitting em modelos de aprendizado de máquina, melhorando a eficácia e eficiência das análises e algoritmos.
