```python
# Modelo no supervisionado de reduo
# Importar as bibliotecas necessrias
import pandas as pd
from sklearn.decomposition import PCA

# Carregar os dados
df = pd.read_csv('creditcard.csv')

# Pr-processar os dados
df = df.drop(['Class'], axis=1)

# Aplicar PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(df)

# Criar DataFrame para os componentes principais
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])

print(principalDf.head())

import matplotlib.pyplot as plt

# Plotar os componentes principais em um grfico de disperso 2D
plt.figure(figsize=(8, 6))
plt.scatter(principalDf['principal component 1'], principalDf['principal component 2'], alpha=0.5)
plt.title('PCA Plot - Componentes Principais')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.grid(True)
plt.show()
```

A Anlise de Componentes Principais (PCA)  uma tcnica estatstica utilizada para reduzir a dimensionalidade dos dados enquanto mantm o mximo possvel de informao. Ao aplicar PCA ao conjunto de dados creditcard.csv, os valores resultantes apresentam as coordenadas das amostras em um novo espao definido pelos componentes principais. Esses componentes principais so combinaes lineares das variveis originais, onde o primeiro componente captura a maior variabilidade dos dados e cada componente subsequente captura a variabilidade restante de forma ortogonal aos anteriores. Portanto, os valores exibidos para "principal component 1" e "principal component 2" representam as coordenadas das amostras nesse novo espao de duas dimenses, proporcionando uma representao mais compacta e informativa dos dados originais.

A reduo de dimensionalidade com PCA  importante porque simplifica a anlise de dados complexos, tornando mais fcil visualizar e interpretar informaes significativas. Ao reduzir os dados a poucas componentes principais, elimina-se o rudo e melhora-se o desempenho computacional, facilitando a identificao de padres e anomalias. Alm disso, ajuda a prevenir overfitting em modelos de aprendizado de mquina, melhorando a eficcia e eficincia das anlises e algoritmos.
